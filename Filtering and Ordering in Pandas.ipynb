### Filtering and Ordering in Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv(r"/Users/darllabulagner/Desktop/SQL THINGS/CSV FILES/Coffee_Chain_Sales .csv")
df
```

```python
# This will display all the rows that has Sales greater than 500

df[df['Sales'] > 500 ] 
```

```python
specific_countries = ['Texas', 'Nevada']

df[df['State'].isin(specific_countries)]
```

```python
# Let's set the index in State

df2 = df.set_index('State')
df2
```

```python
df2 = df2.filter(items = ['State','Cogs'], axis = 1)
df2
```
