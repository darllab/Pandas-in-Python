### Group By and Aggregating

```python
import pandas as pd
```

```python
df = pd.read_csv(r"/Users/darllabulagner/Desktop/SQL THINGS/CSV FILES/Coffee_Chain_Sales .csv")
df
```

```python
group_by_frame = df.groupby('Product_type')
```

```python
group_by_frame.mean()
```

```python
df.groupby('Product_type').mean()
```

```python
df.groupby('Product_type').count()
```

```python
df.groupby('Product_type').sum()
```

```python
df.groupby('Product_type').agg({'Profit': ['mean','max','count','sum'], 'Sales':['mean','max','count','sum'] })
```

```python
df.groupby('Product_type').describe()
```
